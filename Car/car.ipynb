## Lấy dữ liệu từ mỗi url

```python
cols=['url',"Hãng","Năm sản xuất","Nhiên liệu","Kiểu dáng","Dòng xe","Số km đã đi","Hộp số","Xuất Xứ","Số chỗ"]
```

```python
cars=[]
for i in df_url['url']:
    r = requests.get(i)
    html_text = r.text
    soup = BeautifulSoup(html_text)
   
    if soup.find('span',{'itemprop':'carbrand'}) == None:
        company="NaN"
    else:
        company= soup.find('span',{'itemprop':'carbrand'}).text.strip()
    if soup.find('span',{'itemprop':'mfdate'})==None:
        year="NaN"
    else:
        year=soup.find('span',{'itemprop':'mfdate'}).text.strip()
    if soup.find('span',{'itemprop':'fuel'}) == None:
        fuel ="NaN"
    else:
        fuel=soup.find('span',{'itemprop':'fuel'}).text.strip()
    if soup.find('span',{'itemprop':'cartype'})==None:
        cartype="NaN"
    else:
        cartype=soup.find('span',{'itemprop':'cartype'}).text.strip()
    if soup.find('span',{'itemprop':'carmodel'})==None:
        carmodel="NaN"
    else:
        carmodel=soup.find('span',{'itemprop':'carmodel'}).text.strip()
    if soup.find('span',{'itemprop':'mileage_v2'})==None:
        mile="NaN"
    else :
        mile=soup.find('span',{'itemprop':'mileage_v2'}).text.strip()
    if soup.find('span',{'itemprop':'gearbox'})==None:
        gear="NaN"
    else:
        gear=soup.find('span',{'itemprop':'gearbox'}).text.strip()
    if soup.find('span',{'itemprop':'carorigin'})==None:
        origin="NaN"
    else:
        origin=soup.find('span',{'itemprop':'carorigin'}).text.strip()
    if soup.find('span',{'itemprop':'carseats'})==None:
        seat="NaN"
    else:
        seat=soup.find('span',{'itemprop':'carseats'}).text.strip()
    cars.append([i,company,year,fuel,cartype,carmodel,mile,gear,origin,seat])
```

```python
print(cars)
```

```python
with open("dataset.csv", 'wt', encoding='UTF8', newline='') as f:
    writer = csv.writer(f,delimiter='\t')
    writer.writerow(cols)
    writer.writerows(cars)
f.close()
```
